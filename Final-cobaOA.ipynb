```python
database

data_not_found = []

import time

start = time.time()

for row in database.itertuples():
    #print("{} - {}".format(row.airportICAO, row.rwyIdent))
    
    #Ngambil data
    OAData.loc[(OAData["airport_ident"] == row.airportICAO)]
    
    #Ngambil identifier
    if (row.rwyIdent in OAData.loc[(OAData["airport_ident"] == row.airportICAO) & ((OAData["le_ident"] == row.rwyIdent) | (OAData["he_ident"] == row.rwyIdent)), "le_ident"].to_string().split(" ")):
        is_le = True
        identifier = "le_ident"
        
    elif (row.rwyIdent in OAData.loc[(OAData["airport_ident"] == row.airportICAO) & ((OAData["le_ident"] == row.rwyIdent) | (OAData["he_ident"] == row.rwyIdent)), "he_ident"].to_string().split(" ")):
        is_le = False
        identifier = "he_ident"
        
    else:
        identifier = "not_found"
        print("Data not found")
    
    #print(identifier)
    #print(row.airportICAO)
    #print(row.rwyIdent)
    
    if identifier != "not_found":
        #Ekstrak data
        if is_le:
            cursor = "le_heading_degT"
        else:
            cursor = "he_heading_degT"
            (OAData["airport_ident"] == "AYPY") & (OAData["le_ident"] == "14L")
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" "))
        try:
            rtb = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            rtb = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyTrueBearing"] = rtb

        if is_le:
            cursor = "le_elevation_ft"
        else:
            cursor = "he_elevation_ft"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            rte = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            rte = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        rte = float(rte)*0.3048
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyThresElevation"] = rte

        cursor = "width_ft"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            wft = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            wft = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        wft = float(wft)*0.3048
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyWidth"] = wft

        if is_le:
            cursor = "le_longitude_deg"
        else:
            cursor = "he_longitude_deg"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            lon = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            lon = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyLon"] = lon

        if is_le:
            cursor = "le_latitude_deg"
        else:
            cursor = "he_latitude_deg"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            lat = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            lat = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyLat"] = lat

        cursor = "length_ft"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            leng = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            leng = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        leng = float(leng)*0.3048 
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwyLength"] = leng

        cursor = "surface"
        #print(OAData.loc[(OAData["airport_ident"] == row.airportICAO), cursor].to_string().split(" "))
        try:
            sur = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[4]
        except:
            sur = OAData.loc[(OAData["airport_ident"] == row.airportICAO) & (OAData[identifier] == row.rwyIdent), cursor].to_string().split(" ")[3]
        database.loc[(database["airportICAO"] == row.airportICAO) & (database["rwyIdent"] == row.rwyIdent), "rwySurface"] = sur
    
    else:
        data_not_found.append("{}|{}".format(row.airportICAO, row.rwyIdent))
    
database
print(round(time.time()-start,3))
```
